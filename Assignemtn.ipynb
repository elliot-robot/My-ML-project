##For the rest of the variables, applying dummy coding to deal with them

```python
ds = pd.get_dummies(ds,columns=['workclass','occupation','relationship','race','maritalstatus'],dtype=int)
ds.head()
```

```python
#shape of the Dataset
ds.shape
```

#### Define the input variables and the target variable
target variable is the income, and input variables are the rest of the columns.

```python
array=ds.values
X=array[:,1:41]
y=array[:,0]

```

#### Split the training and testing dataset
Randomly sample the dataset with a random state of 123, use 90% for training and 10% for testinguse

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

#### Apply normalization on both train and testing dataset
One reason this is important is because the variables are multiplied by the model weights. So the scale of the outputs and the scale of the gradients are affected by the scale of the inputs.

Although a model might converge without normalization, normalization makes training much more stable.

```python
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing data
X_test_norm = norm.transform(X_test)
```

# Task-2

### Now we are learning how to train a model with logistic regression and SVM for classification, based on entire training dataset and then evaluating the model based on testing dataset
Be aware that, for regression model, the default evaluation metrics is R Squared. For regression task, the default evaluation metrics is accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC


Model1 = LogisticRegression(solver="liblinear")
Model1.fit(X_train_norm, y_train)
test_score = Model1.score(X_test_norm, y_test)
print("Testing Accuracy of LR:", test_score)


Model2 = SVC()
Model2.fit(X_train_norm, y_train)
test_score = Model2.score(X_test_norm, y_test)
print("Testing Accuracy of SVC:", test_score)
```

### Training a model with 10-fold cross valiation

##### Defining a 10 fold cross validation with data shufflling and set the random state with 2

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, shuffle=True, random_state=2)
```

##### Running the 10-fold cross validation and print the average accuracy score based on the cross validation results, and evaluating both model on the testing dataset

```python
from sklearn.model_selection import cross_val_score

Model1 = LogisticRegression(solver="liblinear")
results = cross_val_score(Model1, X_train_norm, y_train, cv=kfold)
print("Average Accuracy of LR:",results.mean())

Model2 = SVC()
results = cross_val_score(Model2, X_train_norm, y_train, cv=kfold)
print("Average Accuracy of SVM:",results.mean())
```

#Applying Parameter Finetuning for two models 

```python
# fine tune parameters for logistic regression model
from sklearn.model_selection import GridSearchCV

grid_params_lr = {
    'penalty': ['l1', 'l2'],
    'C': [1, 10],
    'solver': ['saga', 'liblinear']
}

lr = LogisticRegression(max_iter=90)
gs_lr_result = GridSearchCV(lr, grid_params_lr, cv=kfold).fit(X_train_norm, y_train)
print(gs_lr_result.best_score_)
```

 Evaluate the trained Logistic Regression model using testing dataset

```python
test_accuracy = gs_lr_result.best_estimator_.score(X_test_norm, y_test)
print("Accuracy in testing:", test_accuracy)
```

```python
#checking the parameter setting for the best selected model

gs_lr_result.best_params_
```

```python
#Applying Parameter Finetuning for SVC models 

from sklearn.model_selection import GridSearchCV

grid_params_svc = {
    'kernel': ['linear', 'poly'],
    'C': [1, 10],
    'degree': [3, 8],
    'gamma': ['auto','scale']
}

svc = SVC()
gs_svc_result = GridSearchCV(svc, grid_params_svc, cv=kfold).fit(X_train_norm, y_train)
print(gs_svc_result.best_score_)
```

Evaluate the trained svm model using testing dataset

```python
test_accuracy = gs_svc_result.best_estimator_.score(X_test_norm, y_test)
print("Accuracy in testing:", test_accuracy)
```

```python
from sklearn.cluster import KMeans


X1 = array[:,0:-1]

# fit scaler on training data
norm = MinMaxScaler().fit(X1)
X1 = norm.transform(X1)

kmeans = KMeans(n_clusters=2, random_state=0).fit(X1)
```

```python
#Number of Data Samples in Each Cluster.
from sklearn.preprocessing import MinMaxScaler
import numpy as np

kmeans_labels = kmeans.labels_

unique_labels, unique_counts = np.unique(kmeans_labels, return_counts=True)
dict(zip(unique_labels, unique_counts))
```

Extract prototype for each cluster

```python
from sklearn.metrics.pairwise import pairwise_distances_argmin

kmeans_cluster_centers = kmeans.cluster_centers_
closest = pairwise_distances_argmin(kmeans.cluster_centers_, X1)

# show the two data samples that can represent the two clusters
ds.iloc[closest, :]
```

 Check accuracy of the clustering model

```python
from sklearn.metrics import accuracy_score

y = array[:,0:1]
kmeans_labels = kmeans.labels_


accuracy = accuracy_score(y, kmeans_labels)
print("k means prediction accuracy:", accuracy)
```
